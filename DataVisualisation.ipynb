```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb

data = "/content/HRDataset_v14.csv"
df = pd.read_csv(data)
genderSalary = df.groupby('Sex')['EmpSatisfaction'].mean()
genderSalary.plot(x = 'Sex', y= "EmpSatisfaction", kind="bar", title="Employment Satisfaction between Genders", xlabel="Gender", ylabel="Satisfaction")
plt.show()

"""
We can infer from the graph that both male and females do not have a large difference in
employment satisfaction - most often on average; they're satisfied.
"""
```

```python
"data = "/content/HRDataset_v14.csv"
df = pd.read_csv(data)
positionSalary = df.groupby('Position')['EmpSatisfaction'].mean()
positionSalary.sort_values().plot(x = 'Position', y= "Satisfaction", kind="bar", title="Satisfaction between positions", xlabel="Positions", ylabel="Satisfaction")
plt.show()

"""
Judging from the graph, we can infer that each job has their level of satisfaction
in employment and that 6 of them receive the highest possible employment satisfaction score,
and another 6 receives the lowest and one of them is having the highest paid job.
"""
```

```python
data = "/content/HRDataset_v14.csv"
df = pd.read_csv(data)
positionSalary = df.groupby('MaritalDesc')['EmpSatisfaction'].mean()
positionSalary.sort_values().plot(x = 'Marital Description', y= "Employment Satisfaction", kind="bar", title="Marital Status vs Employment Satisfication", xlabel="Marital", ylabel="Satisfaction")
plt.show()

"""
When analysing the graph between martial status and how much they enjoy their job, we get a
slight idea to infer from - that being if you're without a partner; you're more likely
to enjoy your job. Widowed having a higher score than 4.0 compared to being married that
is sittting just above 3.5.
"""
```

```python
data = "/content/HRDataset_v14.csv"
df = pd.read_csv(data)
positionSalary = df.groupby('EmploymentStatus')['EmpSatisfaction'].mean()
positionSalary.sort_values().plot(x = 'Employment Status', y= "Employment Satisfaction", kind="bar", title="Employment Status vs Employment Satisfication", xlabel="Employment Status", ylabel="Satisfaction")
plt.show()

"""
We can infer from this graph, there's not much of a relationship between your status as an employee to a job
and how much you've enjoyed working there. The difference is too minor to make infer any ideas other than
it has no corrrelation.
"""
```

```python
data = "/content/HRDataset_v14.csv"
df = pd.read_csv(data)
positionSalary = df.groupby('TermReason')['EmpSatisfaction'].mean()
positionSalary.sort_values().plot(x = 'Termination Reasons', y= "Employment Satisfaction", kind="bar", title="Termination Status vs Employment Satisfication", xlabel="Termination Reason", ylabel="Satisfaction")
plt.show()

"""
Extending from the previous graph's inference, this provides a more in-depth analysis to people who has
left their job for a reason. We can see that the lowest employment satisfaction scores are due to
personal issues. Whereas the termination reasons for the higher employment satisfaction scores seem to lean
more towards career changing options.
"""
```

```python
data = "/content/HRDataset_v14.csv"
df = pd.read_csv(data)
positionSalary = df.groupby('Department')['EmpSatisfaction'].mean()
positionSalary.sort_values().plot(x = 'Department', y= "Employment Satisfaction", kind="bar", title="Department vs Employment Satisfication", xlabel="Department", ylabel="Satisfaction")
plt.show()

"""
From the graph produced, we can see the correlation between departments and employment satisfaction. That is,
those that are in an office department are often least enjoying their job. Those who are assgiend to tech departments
are more likely to score higher on employment satisfaction.
"""
```

```python
data = "/content/HRDataset_v14.csv"
df = pd.read_csv(data)
positionSalary = df.groupby('PerformanceScore')['EmpSatisfaction'].mean()
positionSalary.sort_values().plot(x = 'Performance Score', y= "Employment Satisfaction", kind="bar", title="Performance Score vs Employment Satisfication", xlabel="Performance Score", ylabel="Satisfaction")
plt.show()

"""
The graph provides a clear correlations of how performance score of an employee may affect how much
they feel about their job. Those who seem to lack behind in performance in their job enjoy much less,
whereas, those who excels in their job score a 1.5 difference between highest and lowest
"""
```
